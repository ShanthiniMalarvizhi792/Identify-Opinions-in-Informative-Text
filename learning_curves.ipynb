```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_loss_accuracy(train_loss_file, val_loss_file):
    # Load CSV files
    train_loss_data = pd.read_csv(train_loss_file)
    val_loss_data = pd.read_csv(val_loss_file)

    # Extract data
    steps_train = np.array(train_loss_data['Step'])
    train_loss = np.array(train_loss_data['Value'])
    steps_val = np.array(val_loss_data['Step'])
    val_loss = np.array(val_loss_data['Value'])

    # Check lengths of data
    print(f"Training Steps Length: {len(steps_train)}, Validation Steps Length: {len(steps_val)}")
    print(f"Training Loss Length: {len(train_loss)}, Validation Loss Length: {len(val_loss)}")

    # Find common steps
    common_steps = np.intersect1d(steps_train, steps_val)

    # Filter the loss values based on common steps
    train_loss_filtered = train_loss[np.isin(steps_train, common_steps)]
    val_loss_filtered = val_loss[np.isin(steps_val, common_steps)]

    # Make sure the filtered losses are of the same length
    min_length = min(len(train_loss_filtered), len(val_loss_filtered), len(common_steps))
    common_steps = common_steps[:min_length]
    train_loss_filtered = train_loss_filtered[:min_length]
    val_loss_filtered = val_loss_filtered[:min_length]

    # Plotting
    plt.figure(figsize=(10, 5))

    # Plot Training Loss and Validation Loss
    plt.plot(common_steps, train_loss_filtered, label='Training Loss', color='blue')
    plt.plot(common_steps, val_loss_filtered, label='Validation Loss', color='orange')
    plt.title('Training and Validation Loss')
    plt.xlabel('Step')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

def main():
    train_loss_file = './information-model/logs/info_tr_loss.csv'
    val_loss_file = './information-model/logs/info_eval_loss.csv'
    plot_loss_accuracy(train_loss_file, val_loss_file)

if __name__ == "__main__":
    main()
```
